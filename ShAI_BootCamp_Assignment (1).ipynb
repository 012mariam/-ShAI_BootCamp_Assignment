## **Descriptive Statistics**

```python
df.describe()
```

```python
# Calculate mean, median, mode, minimum, and maximum salary
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]  # Note: mode can have multiple values, so we take the first one
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

# Determine the range of salaries
salary_range = max_salary - min_salary

# Find the standard deviation of salaries
std_dev_salary = df['TotalPay'].std()

# Print the results
print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

## **Data Cleaning**

```python

# Drop columns
columns_to_drop = ['Notes', 'Status']
df.drop(columns=columns_to_drop, inplace=True)

```

```python
df
```

```python

columns_to_dropna= ['Benefits']

# Drop rows with NaN values in the specified column(s)
df = df.dropna(subset=columns_to_dropna)

```

```python
df.isna().sum()
```

```python
unique_values = df['BasePay'].unique()
unique_values
```

```python


# Specify the column(s) for which you want to fill missing values
columns_to_fill = ['BasePay']

# Fill missing values with the average value of each column
for column in columns_to_fill:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)


```

## **Simple Correlation Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # for better visualization of scatter plots

# Assuming your DataFrame is named 'df' and has the mentioned columns

# Selecting numerical columns
numerical_columns = df.select_dtypes(include='number')

# Calculating the correlation matrix
correlation_matrix = numerical_columns.corr()

# Displaying the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Plotting scatter plots for each numerical column against 'TotalPay'
for column in numerical_columns.columns:
    if column != 'TotalPay':
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=df[column], y=df['TotalPay'])
        plt.title(f'Scatter Plot: TotalPay vs {column}')
        plt.xlabel(column)
        plt.ylabel('TotalPay')
        plt.show()
```
